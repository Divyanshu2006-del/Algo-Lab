# Travelling Salesman Problem using Dynamic Programming.
Problem Description:
Given a set of cities and distances, find the shortest possible route that visits each city once and returns to the starting city.

#Theory
TSP uses dynamic programming with bit masking. Store results of visited subsets to reduce repeated work.

## Algorithm
1. Start from source city
2. Mark visited cities using bit mask
3. Try all unvisited cities
4. Choose minimum cost path
5. Return to source city

```python
from os import path
def tsp(graph):
    n = len(graph)
    dp = [[-1]*(1<<n) for _ in range(n)]
    parent = [[-1]*(1<<n) for _ in range(n)]

    def visit(city, mask):
        if mask == (1<<n) - 1:
            return graph[city][0]

        if dp[city][mask] != -1:
            return dp[city][mask]

        ans = float('inf')
        best_next_city = -1

        for next_city_idx in range(n):
            if not (mask & (1<<next_city_idx)):
                current_cost = graph[city][next_city_idx] + visit(next_city_idx, mask | (1<<next_city_idx))
                if current_cost < ans:
                    ans = current_cost
                    best_next_city = next_city_idx

        dp[city][mask] = ans
        parent[city][mask] = best_next_city
        return ans

    min_cost = visit(0, 1)

    path = [0]
    mask = 1
    city = 0
    while parent[city][mask] != -1:
        next_city = parent[city][mask]
        path.append(next_city)
        mask |= (1 << next_city)
        city = next_city
    path.append(0)

    return min_cost, path

n = int(input("Enter the number of cities: "))
graph=[]
print("Adjacency Matrix")
for i in range(n):
  row = list(map(int,input().split()))
  graph.append(row)
min_cost, path = tsp(graph)
print("Minimum Travelling Cost: ",min_cost)
print("Path of cities: ",path)
```

```python
graph = [
[0, 10, 15, 20],
[10, 0, 35, 25],
[15, 35, 0, 30],
[20, 25, 30, 0]
]

print("Minimum travelling cost:", tsp(graph))
```
